##### Euclidean Distance Calculator###

/////Objective: Write a function to calculate the Euclidean distance between two vectors///

```python
import numpy as np
v1=np.array([[2,3],[3,4]])
v2=np.array([[3,4],[4,5]])
print("value of v1 is:\n",v1)
print("valuye of v2 is:\n",v2)
```

////Euclidean calculation///

```python
def euclidean_fun(v1,v2):
    sqrt=np.sqrt(np.sum((v1-v2)**2,axis=1))### formula Σ(v1−v2)2.in row wise 
    return sqrt
distance=euclidean_fun(v1,v2)
print("Euclidean calucluation of v1 and v2 is:\n",distance)

```
