////***From Data to Matrix***///

/////Objective: Convert a raw dataset into a NumPy matrix and perform basic transformations////

```python
import numpy as np
import pandas as pd
student_grades=[["Anu",80,54],["priya",57,80],["vishnu",89,45],["ram",66,70]]
arr=np.array(student_grades)
print(arr)
```

//////convert to matrix////

```python
name=np.array([row[0]for row in arr]).reshape(-1,1)
score=[row[1:] for row in arr]
matrix=np.array(score,dtype=float)
print(matrix)
column=["maths","science"]
## using dataframe to display with column name###
df=pd.DataFrame(matrix,columns=column)
print(df)
```

/////Add a new feature(column)///

```python
sports_score=np.array([50,40,56,78]).reshape(-1,1) ## another way sports_score=[[50],[40],[56],[78]]
new_column=np.hstack((matrix,sports_score))
print(new_column)
###using dataframe####
column=["maths","science","sports_score"]
df1=pd.DataFrame(new_column,columns=column,index=[row[0]for row in arr]) ###index value always starts in 0,index(0,4)
print(df1)
```

/////calculating a new feature average/////

```python
average=np.mean(new_column,axis=1).reshape(-1,1)
new_matrix=np.hstack((new_column,average))
print(new_matrix)
```

///final output///


```python

print("Display with Names")
output=np.hstack((name,new_matrix))
print(output)
new_matrix[:, 0] *= 1.1
print("After adding Scaler")
print(new_matrix)
```
