////**Simple Neural Network Layer Simulation**///

////Objective: Simulate a single layer of a neural network using matrix and vector multiplication.///

```python
import numpy as np
input_vector=np.array([2,3,4])
print(F"Input data:{input_vector}")
weight_matrix=np.random.rand(3,2)
bias_vector=np.array([0.1,0.2])
print("weight_matrix\n",weight_matrix)
print("bias vector\n",bias_vector)
###perform linear transformation##
output=input_vector@weight_matrix+bias_vector
print("linear transformation")
print(output)
##sigmoid function##
def sigmoid(x):
    return 1/(1+np.exp(-x))
acti_func=sigmoid(output)
print("non-linear function")
print(acti_func)

```
